```python
# -*- coding: utf-8 -*-
"""
Program used to check what date a specified user has earned the Black Dawn badge.
"""
import requests

while True:
    userID = -1

    while userID < 0:
        try:
            userID = input('Enter a valid Roblox user ID (or enter 0 to quit): ')
            userID = int(userID)

            if userID < 0:
                print('Invalid user ID.')
        except ValueError:
            print('Invalid user ID.')
            userID = -1

    if userID == 0:
        break

    r = requests.get(f'https://badges.roblox.com/v1/users/{userID}/badges/2124422246/awarded-date')
    if r.text == '':
        print(f'User {userID} either does not exist or does not have the Black Dawn badge.')
    elif 'error' in r.text:
        print(f'User {userID} does not exist.')
    else:
        print(f'User ID: {userID}')
        print(f'Date: {r.text[37:47]}')
        print(f'Time: {r.text[48:-3]} UTC')
```
